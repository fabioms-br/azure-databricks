##Ler os arquivos em formato CSV

```python
file2018 = 'wasbs://' + blob_container + '@' + storage_account_name + '.blob.core.windows.net/SUP_IES_2018.CSV'
file2019 = 'wasbs://' + blob_container + '@' + storage_account_name + '.blob.core.windows.net/SUP_IES_2019.CSV'

sup_ies_2018 = spark.read.format('csv').load(file2018, inferSchema = True, header = True, delimiter = '|', encoding = 'cp1252')
sup_ies_2019 = spark.read.format('csv').load(file2019, inferSchema = True, header = True, delimiter = '|', encoding = 'cp1252')
```

##Converter Spark Dataframe para Pandas

```python
df_2018 = sup_ies_2018.toPandas()
df_2019 = sup_ies_2019.toPandas()
```

##Combinar os registros

```python
df = df_2018.append([df_2019])
```
