## Acessar arquivos no Azure Data Lake

```
# Importar Pacotes instalados
import pandas as pd
from azure.storage.filedatalake import DataLakeServiceClient

# Definir caminho do arquivo
arquivo_01 = 'https://fabiomsdatalake.blob.core.windows.net/fabiomsfilesys/linkedin/2022_FabioMenezes_2022_07_03.xlsx'
arquivo_02 = 'https://fabiomsdatalake.blob.core.windows.net/fabiomsfilesys/linkedin/2022_FabioMenezes_2022_08_04.xlsx'

# Obter Token de Acesso
sas_token = '?sv=2021-06-08&ss=bfqt&srt=o&sp=rwdlacupyx&se=2022-10-10T01:01:00Z&st=2022-10-09T17:01:00Z&spr=https&sig=id6XkLfWHMpZ9DBww8H1PJplzREkx9%2Bxy29bUv77T%2F4%3D'

# Ler arquivo excel
data_01 = pd.read_excel(arquivo_01 + sas_token)
data_02 = pd.read_excel(arquivo_02 + sas_token)

# Combinar valores
df = data_01.append([data_02])
```

## Conhecer os dados

```
df.info()
```

## Identificar registros duplicados

```
duplicados = df[df.duplicated()]
duplicados
```

```
duplicados.count()
```

## Remover registros duplicados

```
resultado = df.drop_duplicates()
resultado
```

```
resultado[resultado.duplicated()].count()
```

## Salvar Resultado em Tabela

```
# Converter em Spark Dataframe 
spark_df = spark.createDataFrame(resultado)

# Gravar os registros
spark_df.write.mode("overwrite").saveAsTable("linkedin_analytics")
```
